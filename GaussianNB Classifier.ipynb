```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as mtp
```

```python
data = pd.read_csv(r'C:\Users\hp\Documents\forbidden\Hidden_Databases\NB_Car_Ad.csv')
data
```

```python
x = data.iloc[:,[2,3]].values
y = data.iloc[:,4].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (x,y, test_size = 0.25, random_state = 0)
```

```python
# Feature scaling 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
# Fitting naive bayes to training set
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(x_train, y_train)
```

```python
# Predicting the test set results
y_pred = classifier.predict(x_test)
```

```python
# Making the confusion matrix 
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

As we can see in the above confusion matrix output, there are 5+4 = 9 incorrect prdictions, and 64+27 = 91 correct predictions

```python
# Visualising the Training set results
from matplotlib.colors import ListedColormap

x_set, y_set = x_train, y_train
X1, X2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 1, stop = x_set[:, 0].max() + 1, step = 0.01), np.arange(start = x_set[:, 1].min() - 1, stop = x_set[:, 1].max() + 1, step = 0.01))
mtp.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),alpha= 0.75, cmap = ListedColormap(('yellow', 'purple')))
mtp.xlim(X1.min(), X1.max())
mtp.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)): 
    mtp.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
                c = ListedColormap(('purple', 'yellow')) (i), label = j)
mtp.title('Naive Bayes (Training set)')
mtp.xlabel('Age')
mtp.ylabel('Estimated Salary') 
mtp.legend()
mtp.show()
```
